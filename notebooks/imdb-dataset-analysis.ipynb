Here I'm trying to analyze the [free datasets](https://developer.imdb.com/non-commercial-datasets/) provided by IMDB.

```python
import pandas as pd
```

```python
df_title_basics = pd.read_csv("../data/title.basics.tsv", sep="\t")
df_title_basics
```

```python
df_title_ratings = pd.read_csv("../data/title.ratings.tsv", sep="\t")
df_title_ratings
```

```python
df_title_basics.info()
```

```python
df = pd.merge(df_title_basics, df_title_ratings, how="inner", on="tconst")
df
```

```python
df["titleType"].unique()
```

```python
useful_title_types = ["movie", "tvMovie", "tvSeries", "tvMiniSeries"]
df = df[df["titleType"].isin(useful_title_types)]
df
```

```python
df = df.dropna()
df
```

```python
df["imdb_id"] = df["tconst"].apply(lambda x: int(x.lstrip("t")))
df
```

I want to check, if I can get the top-250 as on IMDB

```python
df = df[df["startYear"] != "\\N"]
df
```

```python
df[(df["numVotes"] > 100000) & (df["titleType"] == "movie")].nlargest(
    250, "averageRating"
)
```

And now I need a list of genres

```python
genres = set()
for genre_list in df[df["genres"] != "\\N"]["genres"].unique():
    genres.update(genre_list.split(","))
genres
```
